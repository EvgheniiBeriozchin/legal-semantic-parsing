```python
import re

conc_regex = re.compile(
    "".join([
        "^Holds that there has been a violation of Article ([0-9]+)( § ?[0-9+])? of the Convention(?:\.)?$"
    ]))

conc_logic = "Court(\\a)^Decide(\\b)^Violation(\\c)^DecidedObject(\\b, \\c)^Article(\\d)^ViolatedObject(\\c, \\d)^ArticleNumber(\\d, {})^ArticleSource(\\d, European Convention of Human Rights)"

count = 0
for item in items_set:
    m = conc_regex.match(item)
    if m is not None:
        count += 1
        groups = m.groups()
        logical_form = conc_logic.format(groups[0])
        if groups[1] is not None:
            logical_form += "^ArticleParagraph(\\d, {})".format(groups[1])
        query = { "text": item }
        update = { "$set": { "logical_form": logical_form } }
        col.update_one(query, update)
        
print("Updated {} items".format(count))
```

```python
def is_item_translated(item):
    cursor = list(col.find({"text": item}))
    if len(cursor) > 0 and "logical_form" in cursor[0]:
        return True
    
percent50 = 0
percent75 = 0
percent100 = 0
less3 = 0

less3_items = []
percent50_items = []
percent75_items = []
percent100_items = []
field = "items_v2"

for i, section in enumerate(all_sections):
    if i % 100 == 0:
        print(i)
    if field not in section or len(section[field]) == 0:
        continue
    translated_count = sum([1 for item in section[field] if is_item_translated(item)])
    if 0 < len(section[field]) - translated_count < 3:
        less3 += 1
        less3_items.append(section[field])
    
    ratio = float(translated_count)/float(len(section[field]))
    if 0.75 > ratio > 0.5:
        percent50 += 1
        percent50_items.append(section[field])
    if 1.0 > ratio > 0.75:
        percent75 += 1
        percent75_items.append(section[field])
    if ratio == 1.0:
        percent100 += 1
        percent100_items.append(section[field])
        
print("Percent50: {}".format(percent50))
print("Percent75: {}".format(percent75))
print("Percent100: {}".format(percent100))
print("Less3: {}".format(less3))
```

```python
to_translate = set()
for section in [item for item in percent50_items + percent75_items if item not in percent100_items]:
    for item in section:
        if not is_item_translated(item) and not item.endswith(":") and not item.endswith(","):
            to_translate.add(item)
```

```python
import re
import copy

conc_regex = re.compile(
    "".join([
        "^Holds that (?:there is no need|it is not necessary) to (?:examine|consider)( separately)?( at this stage)?",
        " the( merits of the)? (applicant(?:'|’)s |applicants(?:'|’)? )?(complaint|(?:remaining |remainder of the )?complaints)",
        "( (?:under|based on)(?:(?: and| taken together with| taken in conjunction with)? Article(?:s)?(?:(?: | (?:and) |, )(?:\d+)(?: §+ \d+)?){1,5}(?: of Protocol No.(?: )?(?:\d+))?(?: (?:of|to) the Convention)?){1,5})?",
        "(?:\.|;)?$"
    ]))

conc_logic = "^".join(["Court(\\a)", "Decide(\\b)", "Must(\\c)", "DecidedObject(\\b, ~\\c)", 
                       "Examine(\\d)", "MustObject(\\c, \\d)",
                      ])

count = 0
parsed_items = []
for item in items_set:
    m = conc_regex.match(item)
    if m is not None:
        count += 1
        
        logical_form = copy.deepcopy(conc_logic)
        groups = m.groups()
        if groups[0] is not None:
            logical_form += "^Separately(\\d)"
        if groups[1] is not None:
            logical_form += "^Timeframe(\\e)^TimeframeEvent(\\e, \\d)^TimeframeType(\\e, now)"
        if groups[4] is not None:
            logical_form += "^Complaint{}(\\f)".format("s" if "complaints" in groups[4] else "")
            if "remainder" in groups[4]:
                logical_form += "^Rest(\\g)^RestObject(\\g, \\f)"
                if groups[2] is not None:
                    logical_form += "Merits(\\h)^Belong(\\h, \\g)^ExaminedObject(\\d, \\h)"
                else:
                    logical_form += "^ExaminedObject(\\d, \\g)"
            else:
                if groups[2] is not None:
                    logical_form += "^Merits(\\h)^Belong(\\h, \\f)^ExaminedObject(\\d, \\h)"
                else:
                    logical_form += "^ExaminedObject(\\d, \\f)"
            if "remaining" in groups[4]:
                logical_form += "Remaining(\\f)"
        else:
            print("ERROR")
            print(item)
            break
        
        if groups[3] is not None:
            logical_form += "^Applicant{}(\\i)^Belong(\\f, \\i)".format("s" if "applicants" in groups[3] else "") 
        
        if groups[5] is not None:
            article_regex = "Article(?:s)?(?: )(?P<art1>\d+)(?: )?(?P<signs>§+)?(?: )?(?P<par1>\d+)?"
            for i in range(2, 12):
                article_regex += "(?: )?(and|,)?(?: )?(?P<obj{}>\d+)?(?: )?(?:§+)?(?: )?(?P<par{}>\d+)?".format(i, i)
            article_regex += "(?:(?: )?of Protocol No.(?: )?(?P<pro>\d+))?(?: (?:of|to) the Convention)?" 
            
            article_part = groups[5]
            search = re.search(article_regex, article_part)
            articles = []
            while search is not None:
                articles.append(search.groupdict())
                article_part = article_part[:search.span()[0]] + article_part[search.span()[1]:]
                search = re.search(article_regex, article_part)
            
            logical_form += "^Convention(\\j)^ConventionName(\\j, European Convention of Human Rights)"
            var = "k"
            
            bigger_one = len(articles) > 1 or (articles[0]["obj2"] is not None and articles[0]["signs"] != '§§')
            if bigger_one:
                logical_form += "^Articles(\\{})".format(var)
                var = chr(ord(var) + 1)
                
            for ag in articles:
                source = "j"
                if ag['pro'] is not None:
                    logical_form += "^Protocol(\\{})^ProtocolNumber(\\{}, {})^Belong(\\{}, \\j)".format(var, var, ag['pro'], var)
                    source = var
                    var = chr(ord(var) + 1)
                    
                logical_form += "^Article(\\{})^ArticleNumber(\\{}, {})^ArticleSource(\\{}, \\{})".format(var, var, ag['art1'], var, source)
                if bigger_one:
                    logical_form += "^MemberOf(\\{}, \\k)".format(var)
                if ag['par1'] is not None:
                    logical_form += "^ArticleParagraph(\\{}, {})".format(var, ag['par1'])

                if ag['obj2'] is not None:
                    k = 2
                    if ag['signs'] is not None and ag['signs'] == '§§':
                        while ag['obj' + str(k)] is not None:
                            logical_form += "^ArticleParagraph(\\{}, {})".format(var, ag['obj' + str(k)])
                            k += 1
                    else:
                        while ag['obj' + str(k)] is not None:
                            var = chr(ord(var) + 1)
                            logical_form += "^Article(\\{})^ArticleNumber(\\{}, {})^ArticleSource(\\{}, \\{})".format(var, var, ag['obj' + str(k)], var, source)
                            if bigger_one:
                                logical_form += "^MemberOf(\\{}, \\k)".format(var)
                            k += 1
                
                var = chr(ord(var) + 1)
                
            logical_form += "^BasedOn(\\b, \\k)"

        query = { "text": item }
        update = { "$set": { "logical_form": logical_form, "type": 4 } }
        col.update_one(query, update)

print("Updated {} items".format(count))
```
