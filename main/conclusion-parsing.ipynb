```python
conclusion_sections = []
non_conclusion_cases = []
for i, case in enumerate(echr_data):
    conclusion = None
    for doc in case["content"]:
        for section in case["content"][doc]:
            if "section_name" in section and section["section_name"] == "conclusion":
                conclusion = section
                
    if conclusion is not None:
        conclusion_sections.append((i, conclusion))
    else:
        non_conclusion_cases.append((i, case["content"]))
```

```python
len(non_conclusion_cases)
```

```python
len(conclusion_sections)
```

```python
sections = []
for index, section in conclusion_sections:
    sentences = []
    to_parse = []
    sentences.append(section["content"])
    to_parse += section["elements"][::-1]
    while len(to_parse) > 0:
        parsed_block = to_parse.pop()
        if parsed_block["content"] != "":
            sentences.append(parsed_block["content"].replace("\xa0", " ").replace("  ", " "))
        if len(parsed_block["elements"]) > 0:
            to_parse += parsed_block["elements"][::-1]
    sections.append((index, " ".join(sentences)))
```

```python
clean_sections = []
for index, section in sections:
    if "done in english" in section.lower():
        clean_sections.append((index, section[:section.lower().find("done in english")]))
    elif "done in french" in section.lower():
        clean_sections.append((index, section[:section.lower().find("done in french")]))
    else:
        clean_sections.append((index, section))
```

```python
len(clean_sections)
```

```python
non_conclusion_cases_concat = []
for index, section in non_conclusion_cases:
    for file in section:
        to_parse += section[file][::-1]
        while len(to_parse) > 0:
            parsed_block = to_parse.pop()
            if parsed_block["content"] != "":
                sentences.append(parsed_block["content"])
            if len(parsed_block["elements"]) > 0:
                to_parse += parsed_block["elements"][::-1]
        non_conclusion_cases_concat.append((index, " ".join(sentences)))
```

```python
non_sections = []
for index, case in non_conclusion_cases_concat:
    start = case.lower().find("for these reasons, the court")
    end = case.lower().find("done in english")
    if end < 0:
        end = case.lower().find("done in french")
        
    non_sections.append((index, case[start:end]))
```

```python
all_sections = sorted(clean_sections + non_sections, key = lambda x: x[0]) 
```

```python
import re
x = '(; )'

delimiters_regex = '(\d+\. )|(i+\. )|(\(i+\) )|(\([a-z]\) )'
def delimiter_type(delimiter):
    if re.compile('\d+\. ').match(delimiter):
        return 0
    if re.compile('i+\. ').match(delimiter):
        return 1
    if re.compile('\(i+\) ').match(delimiter):
        return 2
    if re.compile('\([a-z]\) ').match(delimiter):
        return 3

def extract_items(text, index):
    items = []
    delimiter = re.search(delimiters_regex, text)
    if delimiter is not None:
        span = delimiter.span()
        main_type = delimiter_type(text[span[0]:span[1]])
        if span[0] != 0:
            print("Garbage in section {}: {}".format(index, text[:span[0]]))
            return []
        
        text = text[span[1]:].strip()
        is_previous_main = True
        prefix = ""
        secondary_type = None
        i = 0
        
        next_span = re.search(delimiters_regex, text)
        if next_span is not None:
            next_span = next_span.span()
        while next_span is not None:
            next_type = delimiter_type(text[next_span[0]:next_span[1]])
            item = text[:next_span[0]].strip()
            
            if item.endswith(";"):
                    item = item [:-1]
            subitems = [subitem.strip() for subitem in item.split(";")]
            #print("Iteration {}".format(i))
            #print("Subitems: {}".format(subitems))
            #print("Prefix: {}".format(prefix))
            #print("Previous is main" if is_previous_main else "Previous is secondary")
            #print("Next is main" if next_type == main_type else "Next is not main")
            if is_previous_main:
                if next_type == main_type:
                    items += subitems
                else:
                    if secondary_type is not None and secondary_type != next_type:
                        return []
                
                    if secondary_type is None:
                        secondary_type = next_type
                    
                    items += subitems[:-1]
                    prefix = subitems[-1]
                    if prefix.endswith(":"):
                        prefix = prefix[:-1]
            else:
                if len(subitems) > 1:
                    print("Multiple items in same secondary item at index {}".format(index))
                    return []
                items.append(prefix + " " + subitems[0])
                if next_type == main_type:
                    prefix = ""
                elif secondary_type is not None and secondary_type != next_type:
                        return []
            
            #print("Items: {}".format(items))
            is_previous_main = next_type == main_type
            text = text[next_span[1]:].strip()
            next_span = re.search(delimiters_regex, text)
            if next_span is not None:
                next_span = next_span.span()
            i += 1
                
        if is_previous_main:
            items.append(text)
        else:
            items.append(prefix + text)

            
        return items
    
    else:
        return [subitem.strip() for subitem in text.split(";")]
```

```python
not_extracted_indices = []
for index, section in enumerate(all_sections):
    items = extract_items(section["text"], index)
    if len(items) == 0:
        not_extracted_indices.append(index)
    all_sections[index]["items_v2"] = items
```
